```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def displayImage(img):
    plt.imshow(img, cmap='gray')
    plt.show()
    
    
def displayImages(img1,img2):
    fig= plt.figure(figsize=(10,20))
    fig.add_subplot(1, 2, 1)
    plt.imshow(img1, cmap='gray')
    plt.axis('off')
    plt.title("First")
    
    fig.add_subplot(1, 2, 2)
    plt.imshow(img2, cmap='gray')
    plt.axis('off')
    plt.title("Second")
    plt.show()
```

```python
flower = cv2.imread('../assets/threshold.jpg', 0)

displayImage(flower)
```

```python
def my_threshold(image, threshold, threshold_type):
    img = image.copy()  # Create a copy of the image to preserve the original
    if threshold_type == 'original':
        img[(img < threshold)] = 0
    elif threshold_type == 'invert':
        img[(img >= threshold)] = 0
    else:
        raise ValueError(f"Invalid threshold type: {threshold_type}. Type must be either 'original' or 'invert'.")
        
    return img

flower = cv2.imread('../assets/threshold.jpg', 0)

img1 = my_threshold(flower, 70, 'original')
img2 = my_threshold(flower, 70, 'invert')


displayImages(img1, img2)
```
