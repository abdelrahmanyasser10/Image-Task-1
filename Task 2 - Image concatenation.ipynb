```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread("../assets/concat1.jpg")
image2 = cv2.imread("../assets/concat2.jpg")
image3 = cv2.imread("../assets/concat3.jpg")

cv2.imshow("First", image1)
cv2.imshow("Second", image2)
cv2.imshow("Thrid", image3)
cv2.waitKey(0)
cv2.destroyAllWindows()


def displayImage(img):
    plt.imshow(img, cmap='gray')
    plt.show()

concatenated = cv2.hconcat([image1,image2,image3])
displayImage(concatenated)
```
